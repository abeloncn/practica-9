```python

import numpy as np
matriu1=np.random.randint(0,101,(5,5))
print("Matriu 5x5 amb valors aleatoris \n",matriu1)
print("Valor minim de la matriu: ",matriu1.min())
print("Valor màxim de la matriu: ", matriu1.max())
```

```python
matriu2=np.random.randint(0,101,(10,10))
matriu_normalitzada=(matriu2-matriu2.min())/(matriu2.max()-matriu2.min())
print(matriu_normalitzada)
```

```python
import matplotlib.pyplot as plt
X1=np.linspace(0,2*np.pi,100)
def funcio1(x):
    y=np.cos(2*x-np.pi)/(x+1)
    return y
Y1=funcio1(X1)
plt.plot(X1,Y1,"--",marker="o",markevery=10,markerfacecolor="r",markeredgecolor="r")
plt.xlabel("angle",fontsize=25)
plt.ylabel("$cos(2x-\pi)/(x+1)$",fontsize=15)
plt.title("Exercici 2",fontsize=15)
plt.xlim(0,2*np.pi)
plt.ylim(-1,1)
punt6=np.pi
plt.annotate("Sisè punt",xy=(punt6,funcio1(punt6)),xytext=(2,-0.5),
             arrowprops=dict(facecolor="black"))
plt.show()
```

```python
import scipy.optimize as so
import math
def funcio3(x):
    y=x**2-x
    return y
X3=np.linspace(-3,3,100)
Y3=funcio3(X3)

plt.plot(X3,Y3)

plt.title("Arrel de $x^2-x$")
plt.annotate("Arrel",xy=(so.newton(funcio3,2),0),xytext=(0,6),
             arrowprops=dict(facecolor="black"))
plt.show()
print("La arrel d'aquesta funció es torba a x=",so.newton(funcio3, 4))
def funcio4(x):
    y=np.cos(x)-x
    return y
X4=np.linspace(-3,3,100)
Y4=funcio4(X4)

plt.plot(X4,Y4)

plt.title("Arrel de $cos(x)-x$")
plt.annotate("Arrel",xy=(so.newton(funcio4,2),0),xytext=(-1,-2),
             arrowprops=dict(facecolor="black"))
plt.show()
print("La arrel d'aquesta funció es troba a x=",so.newton(funcio4, 4))
def funcio5(x):
    y=np.sin(x)**2-np.sin(x)
    return y
X5=np.linspace(0,6,100)
Y5=funcio5(X5)

plt.plot(X5,Y5)

plt.title("Arrel de $sin^2(x)-sin(x)$")
plt.annotate("Arrel",xy=(so.newton(funcio5,3),0),xytext=(0.5,1),
             arrowprops=dict(facecolor="black"))
plt.show()
print("La arrel d'aquesta funció es troba a x=",so.newton(funcio5, 3))
```
